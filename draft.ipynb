```python
#pair stats
def get_stats(ids):
    count = {}
    for pair in zip(ids, ids[1:]):
        count[pair] = count.get(pair,0) + 1
    return count

pairs = get_stats(tokens)
print(sorted(((v,k) for k,v in pairs.items()), reverse = True))
```

```python
top_pair = max(pairs, key = pairs.get)
top_pair
```
